```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
# Function to upload a file
def upload_file():
    uploaded = files.upload()
    for filename in uploaded.keys():
        print(f"Uploaded file: {filename}")
        if filename.endswith('.csv'):
            return pd.read_csv(filename)
        elif filename.endswith('.xlsx'):
            return pd.read_excel(filename)
        else:
            print("Unsupported file format. Please upload a CSV or Excel file.")
            return None
```

```python
# Function to summarize the dataset
def dataset_summary(df):
    print("\nDataset Summary:")
    print(df.info())
    print("\nFirst few rows of the dataset:")
    print(df.head())
    print("\nMissing values in each column:")
    print(df.isnull().sum())
```

```python
# Function to visualize sales trends by category
def visualize_sales_trends(df):
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Category', y='Sales', data=df, estimator=sum, ci=None)
    plt.title("Total Sales by Category")
    plt.xlabel("Category")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.barplot(x='Sub Category', y='Sales', data=df, estimator=sum, ci=None)
    plt.title("Total Sales by Sub Category")
    plt.xlabel("Sub Category")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Function to analyze profit vs. discount
def profit_vs_discount(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Discount', y='Profit', hue='Category', data=df, alpha=0.7)
    plt.title("Profit vs Discount by Category")
    plt.xlabel("Discount")
    plt.ylabel("Profit")
    plt.legend(title="Category")
    plt.show()
```

```python
# Main program
print("Upload your Supermart Grocery Orders dataset:")
data = upload_file()

if data is not None:
    # Summary and basic cleaning
    dataset_summary(data)

    # Visualizations
    visualize_sales_trends(data)
    profit_vs_discount(data)

    # Save the cleaned and summarized dataset
    output_file = 'processed_supermart_orders.csv'
    data.to_csv(output_file, index=False)
    print(f"Processed dataset saved as {output_file}")
    files.download(output_file)
```
